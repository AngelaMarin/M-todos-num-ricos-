```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
```

```python
#Definimos una función para calcular el polinomio de Lagrange
def PIL(x,f):
  n=len(x)
  p=sp.symbols('X')
  polinomio=0

  for i in range(n):
    prod=1
    for j in range(n):
      if i!=j:
        prod*=(p-x[j])/(x[i]-x[j])
    polinomio+=f[i]*prod
  return polinomio
```

```python
#Damos los valores para calcular la interpolación:
x=np.array([0,1,2,3,4,5])
f=np.array([-5,1,9,25,55,105])

polinomio=PIL(x,f)
print("El polinomio de interpolación de Lagrange para los datos dados es:",polinomio)
```

```python
#Graficamos
p=sp.symbols('X')
y=sp.lambdify(p,polinomio,'numpy')
x_values=np.linspace(0,5,20)
y_values=y(x_values)

plt.plot(x_values, y_values, label='Polinomio de interpolación de Lagrage')
plt.scatter(x,f, color='black', label='Puntos originales')
plt.xlabel('x')
plt.ylabel('f')
plt.title('Polinomio de interpolación de Lagrange')
plt.legend()
plt.show()
```
