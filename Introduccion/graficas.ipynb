# Gr치fica de $f(x)=x - 3^{-x}$

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Funci칩n cuadr치tica.
def f(x):
    return x - 3**(-x)
def g(x):
    return x
def h(x):
    return 3**(-x)
```

```python
# Valores del eje X que toma el gr치fico.
x=np.linspace(-10, 10,100)
```

```python
# Graficar ambas funciones.
plt.figure(figsize=(25,15))
plt.plot(x, f(x), color="blue" )
plt.plot(x, g(x), color="red")
plt.plot(x, h(x), color="brown")
plt.xlabel("Eje $x$", color="black")
plt.ylabel("Eje $y$", color="black")
plt.grid(color="gray")
plt.axhline(0, color="black")
plt.axvline(0, color="black")
plt.title("Gr치fica de $f(x)=x-3^{-x}$", color="black")
plt.xlim(-10, 10)
plt.ylim(-5, 5)
plt.savefig("output.png") # Guardar gr치fico como im치gen PNG.
# Mostrarlo.
plt.show()
```
