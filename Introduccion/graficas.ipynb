# Gr치fica de $f(x)=e^{x/2}sen(x/3)$

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Funci칩n cuadr치tica.
def f(x):
    return np.exp(x/2)*np.sin(x/3)
```

```python
# Valores del eje X que toma el gr치fico.
x=np.linspace(-100, 100,100000)
```

```python
# Graficar ambas funciones.
plt.plot(x, f(x))
plt.axhline(0, color="black")
plt.axvline(0, color="black")
plt.xlim(-1, 100)
plt.ylim(-100, 100)
#plt.savefig("output.png") # Guardar gr치fico como im치gen PNG.
# Mostrarlo.
plt.show()
```
