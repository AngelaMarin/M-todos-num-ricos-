```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Vamos a definir la función que calculará el polinomio
def PIN(x,f):
  n=len(x)
  p=sp.symbols('X')
  polinomio=f[0]
  for i in range(1,n):
    prod=1
    for j in range(i):
      prod*=(p-x[j])
    polinomio+=f[i]*prod
  return polinomio
```

```python
#Damos los valores para calcular la interpolación:
x=np.array([0,1,2,3,4,5])
f=np.array([-5,1,9,25,55,105])

polinomio=PIN(x,f)
print("El polinomio de interpolación de Newton para los datos dados es:",polinomio)
```

```python
#Graficamos
p=sp.symbols('X')
y=sp.lambdify(p,polinomio,'numpy')
x_values=np.linspace(0,5,20)
y_values=y(x_values)

plt.plot(x_values, y_values, label='Polinomio de interpolación de Newton')
plt.scatter(x,f, color='black', label='Puntos originales')
plt.xlabel('x')
plt.ylabel('f')
plt.title('Polinomio de interpolación de Newton')
plt.legend()
plt.show()
```
