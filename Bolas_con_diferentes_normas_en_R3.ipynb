```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Definimos las normas
#Para la norma p
def normap(x,p):
  return(np.abs(x[0])**p+np.abs(x[1])**p+np.abs(x[2])**p)**(1/p)
#Para la norma infinito
def normainf(x, p):
  #return(np.max(np.abs(x[0]),np.abs(x[1]),np.abs(x[2])))
  return np.max(np.abs(x))

#Definimos una funci√≥n para graficar
def bola(norma, p=None):
  fig=plt.figure(figsize=(5,5))
  ax = fig.add_subplot(111, projection='3d')

  u=np.linspace(0, 2*np.pi, 50)
  v=np.linspace(0, np.pi, 50)
  x=np.outer(np.cos(u), np.sin(v))
  y=np.outer(np.sin(u), np.sin(v))
  z=np.outer(np.ones(np.size(u)), np.cos(v))

  puntos=np.array([x.flatten(), y.flatten(), z.flatten()])
  norm=np.apply_along_axis(lambda vec: norma(vec, p) if p is not None else norma(vec), 0, puntos)
  puntos /=norm

  ax.plot_surface(puntos[0].reshape(x.shape), puntos[1].reshape(y.shape), puntos[2].reshape(z.shape), color='c', alpha=0.6)
  ax.set_title(f"Bola de norma {p}")
  ax.set_xlabel('X')
  ax.set_ylabel('Y')
  ax.set_zlabel('Z')
  plt.show()
```

```python
#Graficamos
bola(normap, 1)
bola(normap, 3/2)
bola(normap, 2)
bola(normap, 5)
bola(normap, 7)
bola(normainf, 'infinito')
```
