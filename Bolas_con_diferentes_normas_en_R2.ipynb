```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Para R2
#Definimos las normas
#Norma p (sirve para los incisos 1, 2, 3, 4 y 5)
def normap(x, p):
  return (np.abs(x[0])**p+np.abs(x[1])**p)**(1/p)
#Norma infinito (inciso 6)
def normainf(x, p):
  return max(np.abs(x[0]), np.abs(x[1]))

#Definimos una funci√≥n para graficar cada norma
def bola(norma, p=None):
  x=np.linspace(-1.2, 1.2, 100)
  y=np.linspace(-1.2, 1.2, 100)
  X, Y=np.meshgrid(x, y)
  Z=np.zeros(X.shape)
  for i in range(X.shape[0]):
    for j in range(X.shape[1]):
      if p is not None:
        Z[i,j]=norma([X[i,j], Y[i,j]], p)
      else:
        Z[i,j]=norma([X[i,j], Y[i,j]])
  plt.figure(figsize=(6,6))
  plt.contourf(X, Y, Z, levels=[0,1], colors='skyblue')
  plt.title(f"Bola con norma {p}")
  plt.axhline(y=0, color='black')
  plt.axvline(x=0, color='black')
  plt.show()
```

```python
#Graficamos ahora cada inciso
bola(normap, 1)
bola(normap, 3/2)
bola(normap, 2)
bola(normap, 5)
bola(normap, 7)
bola(normainf, "infinito")
```
